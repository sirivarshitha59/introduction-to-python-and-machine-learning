Numpy: Numerical Python
Numpy is a general-purpose array-processing package. It provides a high-performance multidimensional array object, and tools for working with these arrays. It is the fundamental package for scientific computing with Python.

Besides its obvious scientific uses, Numpy can also be used as an efficient multi-dimensional container of generic data

Why learn Numpy
Learning NumPy is crucial because it is the foundational library for numerical computing in Python, providing efficient, high-performance, multi-dimensional array objects (called ndarray) and a wide range of mathematical functions to operate on them.

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('variable x')
plt.ylabel('Trigonometric function')
plt.title('Sin & Cos Function')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin, color='green')
plt.title('Sine Function')
plt.xlabel('variable x')
plt.ylabel('Trigonometric function')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine Function')
plt.xlabel('variable x')
plt.ylabel('Trigonometric function')
plt.show()
```

```python
p = np.arange(0,100,5)
q = p**2
d = np.exp(p)
plt.scatter(p,q)
plt.scatter(p,d/1000)
```

```python


# Define y variables using the existing x array
y_linear = x
y_power = x**2
y_exp = np.exp(x)
y_log = np.log(x + 1e-9) # Add a small value to avoid log(0)

plt.figure(figsize=(10,8))

#1. Linear Function

plt.subplot(2,2,1)

plt.plot(x, y_linear, color='blue')

plt.title("Linear Function")

plt.grid()

#2. Power Function

plt.subplot(2,2,2)

plt.plot(x, y_power, color='green')

plt.title("Power Function")

plt.grid()

#3. Exponential Function

plt.subplot(2,2,3)

plt.plot(x, y_exp, color='red')

plt.title("Exponential Function")

plt.grid()

#4. Logarithmic Function

plt.subplot(2,2,4)

plt.plot(x, y_log, color='purple')

plt.title("Logarithmic Function")

plt.grid()

plt.tight_layout()

plt.show()
```
