container types in python
The four core built-in container types are:
list: A mutable, ordered sequence of elements. Lists are indexed by integers and can hold items of different types. They are defined using square brackets (e.g., [1, "a", 2.5]).
tuple: An immutable, ordered sequence of elements. Tuples are similar to lists but cannot be changed after creation. They are defined using parentheses (e.g., (1, "a", 2.5)).
dict (Dictionary): A mutable mapping type that stores data as unordered (ordered since Python 3.6) key-value pairs. Data is accessed using unique keys, not integer indices. They are defined using curly braces (e.g., {"name": "John", "age": 30}).
set: A mutable, unordered collection of unique elements. Sets are useful for operations like union and intersection and for quickly checking membership. They are defined using curly braces (e.g., {1, 2, 3}) or the set() constructor.

```python
# Problem 1: Confusion Matrix Metrics
# Using Python as a Calculator

# Assigning the given values
TP = 50   # True Positives
FP = 8    # False Positives
TN = 32   # True Negatives
FN = 10   # False Negatives

# Calculate total predictions
total = TP + FP + TN + FN

# Calculate Accuracy
accuracy = (TP + TN) / total

# Calculate Precision
precision = TP / (TP + FP)

# Calculate Recall
recall = TP / (TP + FN)

# Calculate F1 Score
f1 = 2 * precision * recall / (precision + recall)

# Print results with clear labels
print("Total Predictions =", total)
print("Accuracy =", accuracy)
print("Precision =", precision)
print("Recall =", recall)
print("F1 Score =", f1)
```

In this code, we used Python to calculate how well a flood prediction model works. We stored the given values, calculated the total predictions, and then used simple formulas to find Accuracy, Precision, Recall, and the F1 Score. Finally, we printed the results to clearly see the model’s performance.

```python
# Problem 2: R² (Coefficient of Determination)
# Using Python as a Calculator

# Given values
SSR = 420   # Sum of Squares due to Regression
SSE = 180   # Sum of Squares due to Error

# Calculate Total Sum of Squares
SST = SSR + SSE

# Calculate R²
R_squared = SSR / SST

# Print results
print("Total Sum of Squares (SST) =", SST)
print("R² =", R_squared)

# Interpretation
if R_squared > 0.7:
    print("The model explains MORE than 70% of the variance.")
elif R_squared == 0.7:
    print("The model explains EXACTLY 70% of the variance.")
else:
    print("The model explains LESS than 70% of the variance.")
```

We added the regression and error sums of squares to get the total variation (SST). Then, by dividing SSR by SST, we calculated R², which shows how well the model fits the data. With R² = 0.7, the model explains 70% of the changes in river discharge

```python
# Problem 3: Mean Squared Error (MSE)
# Using Python as a Calculator

# Step 1: Store squared errors in a list
squared_errors = [2, 4, 1, 9, 4]

# Step 2: Calculate total squared error
total_squared_error = sum(squared_errors)  # sum of all elements

# Step 3: Number of observations
n = len(squared_errors)

# Step 4: Calculate Mean Squared Error (MSE)
MSE = total_squared_error / n

# Step 5: Print the result
print("Mean Squared Error (MSE) =", MSE)
```


The mean, median, and mode are measures that help us understand the central tendency of the fruit harvest data. The **mean** gives the average number of fruits harvested, showing the overall typical value. The **median** represents the middle value when all harvest numbers are arranged in order, making it useful when the data has extreme values. The **mode** shows the number that occurs most frequently, helping us identify the most common harvest amount. Together, these three measures give a clear summary of the dataset.

```python
# Problem 4: Cook's Distance
# Using Python as a Calculator

# Given values
e2 = 9       # Squared residual e_i^2
MSE = 3      # Mean Squared Error
p = 3        # Number of predictors
h_ii = 0.2   # Leverage value
n = 25       # Sample size

# Step 1: Calculate Cook's Distance
D_i = (e2 / (p * MSE)) * (h_ii / (1 - h_ii)**2)

# Step 2: Calculate threshold
threshold = 4 / n

# Step 3: Print Cook's Distance and threshold
print("Cook's Distance (D_i) =", D_i)
print("Threshold (4/n) =", threshold)

# Step 4: Decide if influential
if D_i > threshold:
    print("The observation is INFLUENTIAL.")
else:
    print("The observation is NOT influential.")
```

```python
# Problem 5: Simple Linear Regression Prediction
# Using Python as a Calculator

# Given values
x = 12        # Rainfall
slope = 2.5   # Coefficient of x
intercept = 10  # Intercept

# Step 1: Predict y using the regression equation
y_pred = slope * x + intercept

# Step 2: Print predicted value
print("Predicted crop yield (y) =", y_pred)

# Step 3: Explanation of slope and intercept
print("\nExplanation:")
print("Slope (2.5) means that for every 1 unit increase in rainfall, crop yield increases by 2.5 units.")
print("Intercept (10) means that when rainfall is 0, the predicted crop yield is 10 units.")
```
